## 4.&nbsp;Data Transformation

Now that we have identified all of our data quality issues, our next step is to manipulate the dataset to address each of the issues we found in turn.

### 4.2 students.csv

    a. duplicated rows,
    b. null values within the 'gender' variable, and
    c. value/formatting inconsistencies in 'gender' for e.g. where gender is shown as 'female', 'F', 'Female'..

```python
#Dropping duplicated rows:
df2.drop_duplicates(inplace=True)
df2.duplicated().sum()
```

```python
df2[df2['gender'].isnull()]
```

```python
#Replacing null values in ['gender'] with 'Other'
df2['gender'].fillna('Other', inplace=True)
df2.isnull().sum()
```

```python
#Fixing gender label formatting issue by creating a dictionary
Dict = {'M':'Male', 'F':'Female', 'male':'Male', 'female':'Female'}
df2['gender'].replace(Dict, inplace=True)
df2['gender'].value_counts()
```

### 4.3 enrolments.csv

    a. formatting inconsistency in 'enrolled_year_level' where integers entered in alphabet, and
    b. formatting inconsistency in 'enrolled_class', where the same class type was entered with different formatting

```python
#Change all entries in 'enrolled_year_level' to consistent int64 format by creating a dictionary.
Dict = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9, 'ten':10, 'eleven':11, 'twelve':12, 'zero':0}
df3['enrolled_year_level'].replace(Dict, inplace=True)
df3['enrolled_year_level'] = pd.to_numeric(df3['enrolled_year_level'])
df3['enrolled_year_level'].value_counts().sort_index()
```

```python
class_map = {'A': 'Class_A', 'B': 'Class_B', 'C': 'Class_C', 'D': 'Class_D', 'E': 'Class_E'}
df3['enrolled_class'].replace(class_map, inplace=True)
df3['enrolled_class'].value_counts().sort_index()
```

### 4.5 ae_assessment.csv

    a. formatting inconsistency in 'term_when_assessed' - some entries are a string of alphabet and numbers, but could be an int64 (e.g. 'Term 1' instead of '1')
    b. 'result' contains 1314 missing values

```python
df5['term_when_assessed'].value_counts()
```

```python
#Resolving formatting inconsistency in 'term_when_assessed' by creating a dictionary
term_dict = {'Term 1': 1, 'Term 2': 2, 'Term 3': 3, 'Term 4': 4}
df5['term_when_assessed'].replace(term_dict, inplace=True)
df5['term_when_assessed'] = pd.to_numeric(df5['term_when_assessed'])
df5['term_when_assessed'].value_counts().sort_index()
```

```python
#Filling in 1314 NaN values in ['result'] with 'not assessed'
df5['result'].fillna('not assessed', inplace=True)
df5.isnull().sum()
```

```python
df5['result'].value_counts()
```

```python
# prompt: Use dictionary to replace all results A to E with integers 5 to 1

# Create a dictionary to map letter grades to numerical values
grade_mapping = {'A': 5, 'B': 4, 'C': 3, 'D': 2, 'E': 1}

# Apply the mapping to the 'result' column in df5
df5['result'] = df5['result'].replace(grade_mapping)

# Display the updated 'result' column to verify the changes
print(df5['result'].value_counts())
```

### 4.7 pat_assessment.csv

    a. 25 entries in 'assessment_name' and 'year' with null values

## 6.&nbsp;Pivoting

In this step we will generate a pivot table to show **percentage of different genders enrolled at 'Kirimpika School' in secondary year levels 8, 9, 10 for the year 2020**.

```python
#The required datapoints exist across different sets of data, so we first need to perform a join on the relevant sets.
df_pivot = pd.merge(df2, df3, on=['student_id'])
```

```python
df_pivot.head()
#Merge is successful. I can disregard NaN values because I do not require those columns.
```

```python
#Now generating a pivot table to show % genders enrolled at 'enrolled_school_id' = 4; at enrolled_year_level = 8, 9, 10; for year_of_enrolment = 2021.
df_pivot_filtered = df_pivot[df_pivot['year_of_enrolment'] == 2020]
df_pivot_filtered = df_pivot_filtered[df_pivot_filtered['enrolled_school_id'] == 4]
df_pivot_filtered = df_pivot_filtered[df_pivot_filtered['enrolled_year_level'].isin([8, 9, 10])]

pivot_table = df_pivot_filtered.pivot_table(index='enrolled_year_level', columns='gender', aggfunc='size', fill_value=0)

pivot_table = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

pivot_table = pivot_table.round(1)

pivot_table.columns.name = 'Gender %'

pivot_table
```

```python
#Savings my pivot table to a csv output
pivot_table.to_csv('T2_pivot.csv')
```
